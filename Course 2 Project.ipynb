```python
import pandas as pd
import numpy as np
df = pd.read_csv("course_2_dataset.csv")
df.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
for column in ['tpep_pickup_datetime','tpep_dropoff_datetime']:
    df[column] = pd.to_datetime(df[column])
df['is_store_and_fwd'] = df['store_and_fwd_flag'].apply(lambda x: 1 if x == 'Y' else 0)
df['pickup_day_of_week'] = df['tpep_pickup_datetime'].apply(lambda x: x.dayofweek)
df['pickup_time_of_day'] = df['tpep_pickup_datetime'].apply(lambda x: '00-03' if x.hour < 4 else '04-07' if x.hour < 8 else '08-11' if x.hour < 12 else '12-15' if x.hour < 16 else '16-19' if x.hour < 20 else '20-23')
df['duration_diff'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
df['duration_seconds'] = df['duration_diff'].dt.total_seconds()
df.drop(['store_and_fwd_flag', 'duration_diff'], axis='columns', inplace=True)
df.describe()
```

### Data Observations

- id is unique
- drop duplicates returns the same number of columns when ignoring the id column
- VendorID stands for: 1= Creative Mobile Technologies, LLC; 2= VeriFone Inc.
- RateCodeID is 1= Standard rate, 2=JFK, 3=Newark, 4=Nassau or Westchester, 5=Negotiated fare, 6=Group ride
- PULocationID is the start location id
- DOLocationID is the end location id
- All that are hidden when predicting the duration of the trip because they are known afterward: Trip distance, fare amount, extra, tax, surcharge, tip, total amount, tolls amount.  Also trip distance.
- However, trip distance will be useful for data analysis, but it cannot be used directly for predictions since it's unknown when the ride commenses
- day of week, time of day, and duration needed to be calculated

### Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
table = pd.pivot_table(df, values=['duration_seconds'], index=['pickup_day_of_week'],
                       columns=['pickup_time_of_day'], aggfunc=np.mean)
sns.heatmap(table)
plt.xlabel("Hour of the day")
plt.gca().set_xticklabels([label.get_text()[-5:] for label in plt.gca().get_xticklabels()])
plt.xticks(rotation=90)
plt.title('Ride duration based on day of week and time of day')
```

### Location Distribution

```python
temp = df.groupby(['PULocationID', 'DOLocationID'])['duration_seconds'].agg(['count', 'mean']).reset_index()
temp = temp[temp['count'] > 30]
temp1 = temp.copy()
temp2 = temp.copy()
temp2['PULocationID'] = temp['DOLocationID']
temp2['DOLocationID'] = temp['PULocationID']
# new_temp = pd.concat([temp1, temp2], axis='index', ignore_index=True)
# Realized probably shouldn't duplicate rows
new_temp = temp1.copy()
new_temp = new_temp.drop_duplicates(subset=['PULocationID', 'DOLocationID'])
new_temp = new_temp[new_temp['PULocationID'] != new_temp['DOLocationID']]
new_temp = new_temp.sort_values(['mean'])
new_temp.rename(columns={'mean': 'duration time seconds'}, inplace=True)
new_temp

def concat_columns(x,y):
    return '{}->{}'.format(x,y)
new_temp['key'] = df[['PULocationID', 'DOLocationID']].apply(lambda x: concat_columns(*x), axis='columns').astype(str)
new_temp = new_temp.sort_values(['duration time seconds'])
new_temp = new_temp.iloc[0:]
new_temp['key'] = range(new_temp.shape[0]) 

ax = sns.barplot(data=new_temp, x='key', y='duration time seconds', ci=None)
plt.xlabel("Common start and end locations")
ax.set_xticks([])
plt.title('Ride duration based on 90 most common routes')
```
